```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

# get data from NYTIMES csv
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv'
df = pd.read_csv(url)

# filter NY and take date/case cols
filter_indices = df['state'] == 'New York' 
df = df[filter_indices][['cases']]

# adjust figure size
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

# take the difference between each row to get number of new cases per data
Y = df['cases'].diff().fillna(0).tolist()
X = range(len(Y))

Y = np.log([0.1 if x == 0 else x for x in Y])

print(Y)

# plot it
plt.bar(X, Y)
```
